```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
```

```python

df = pd.read_csv('Topic_Survey_Assignment.csv',index_col = 0) 
df
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt 
mpl.style.use('ggplot') # optional: for ggplot-like style

# check for latest version of Matplotlib
print ('Matplotlib version: ', mpl.__version__) # >= 2.0.0



```

```python
 df.sort_values(['Very interested'], ascending = False , axis=0 , inplace = True)
df = (df*100)/2233  
df = round(df,2)
#  df = round (( df/2233 ) * 100 , 2)
df.head()

```

```python
 x = df.plot(kind='bar',  figsize=(20, 8) , rot = 90 , color=['#5cb85c', '#5bc0de', '#d9534f'], width=.8, fontsize=14)

x.set_facecolor('white')
plt.legend(fontsize=14,facecolor = 'white') 

plt.title("percentage of Respondents' Interest In Data Science Area ", fontsize=16)
x.get_yaxis().set_visible(False)

for p in x.patches:
    x.annotate(np.round(p.get_height(),decimals=2), 
                (p.get_x()+p.get_width()/2., p.get_height()), 
                ha='center', 
                va='center', 
                xytext=(0, 10), 
                textcoords='offset points',
                fontsize = 14
               )


plt.show()
```
